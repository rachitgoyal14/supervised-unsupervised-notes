```python
import numpy as np
import matplotlib.pyplot as plt

# Reproducibility
np.random.seed(42)

# Generate synthetic data
X = 2 * np.random.rand(42, 1)          # 42 samples, 1 feature
y = 4 + 3 * X + np.random.randn(42, 1) # noise added

# Initialize parameters
slope = np.random.randn()
intercept = np.random.randn()

lr = 0.01
epochs = 20

slope_path_sgd = []
intercept_path_sgd = []
```

```python
for epoch in range(epochs):
    for i in range(len(X)):

        rand_i = np.random.randint(len(X))
        xi = X[rand_i]
        yi = y[rand_i]

        # Prediction
        y_pred = slope * xi + intercept
        error = y_pred - yi

        # Gradients
        slope_grad = 2 * error * xi
        intercept_grad = 2 * error

        # Update parameters
        slope -= lr * slope_grad.item()
        intercept -= lr * intercept_grad.item()

        # Store path
        slope_path_sgd.append(slope)
        intercept_path_sgd.append(intercept)


print("Final slope:", slope)
print("Final intercept:", intercept)
```

```python
plt.scatter(X, y, label="Data")
plt.plot(X, slope * X + intercept, color='red', label="SGD Fit")
plt.legend()
plt.title("Linear Regression using SGD")
plt.show()
```

```python
plt.plot(slope_path_sgd, label="Slope Path")
plt.plot(intercept_path_sgd, label="Intercept Path")
plt.legend()
plt.title("Parameter Updates During SGD")
plt.show()
```
