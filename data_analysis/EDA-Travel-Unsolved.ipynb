#### __Separate the categorical and numerical type columns__

#### __Change all the categorical columns to object type__

```python
cat_cols = df.select_dtypes(include=['category']).columns
df[cat_cols] = df[cat_cols].astype('object')
```

#### __Univariante analysis, for all numerical columns using histogram with KDE__

```python
for col in numerical_cols.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram with KDE for {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

#### __Do the same for all categorical columns using countplot__

```python
for col in categorical_cols.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=df[col])
    plt.title(f'Countplot for {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

#### __Perform bivariate analysis on Age and DurationOfPitch such that grouping should be done on ProdTaken__

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(
    data=df,
    x='Age',
    y='DurationOfPitch',
    hue='ProdTaken'
)
plt.title('Bivariate Analysis: Age vs DurationOfPitch grouped by ProdTaken')
plt.show()
```

#### __Draw a cross-tab for rows MaritalStatus and ProdTaken, normalize the same__

```python
pd.crosstab(df['MaritalStatus'], df['ProdTaken'], normalize='index')
```

#### __Draw a barchart that using the data generated above,  The product taken should be lightcoral in color but not taken should be skyblue__

```python
ct = pd.crosstab(
    df['MaritalStatus'],
    df['ProdTaken'],
    normalize='index'
)


ct.plot(
    kind='bar',
    color={1: 'lightcoral', 0: 'skyblue'},
    figsize=(7, 5)
)

plt.title('Product Taken vs Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Proportion')
plt.legend(['Not Taken', 'Taken'])
plt.xticks(rotation=0)
plt.show()
```

#### __Draw a barchart for ProductPitched on X-axis and PitchSatisfactionScore on Y-axism use palatte = 'coolwarm'__

```python
plt.figure(figsize=(7, 5))
sns.barplot(
    data=df,
    x='ProductPitched',
    y='PitchSatisfactionScore',
    palette='coolwarm'
)

plt.title('Average Pitch Satisfaction Score by Product Pitched')
plt.xlabel('Product Pitched')
plt.ylabel('Pitch Satisfaction Score')
plt.xticks(rotation=45)
plt.show()
```

#### __Draw a lineplot for NumberOfFollowups on X-axis and PitchSatisfactionScore on Y-axis, use color = 'purple' and marker = 'o'__

```python
plt.figure(figsize=(7, 5))
sns.lineplot(
    data=df,
    x='NumberOfFollowups',
    y='PitchSatisfactionScore',
    color='purple',
    marker='o'
)

plt.title('Pitch Satisfaction Score vs Number of Follow-ups')
plt.xlabel('Number of Follow-ups')
plt.ylabel('Pitch Satisfaction Score')
plt.show()
```

#### __Draw a violine plot for ProdTaken on X-axis and Age on Y-axis, group them using OwnCar, use palette = 'set1'__

```python
plt.figure(figsize=(7, 5))
sns.violinplot(
    data=df,
    x='ProdTaken',
    y='Age',
    hue='OwnCar',
    palette='Set1'
)

plt.title('Age Distribution by Product Taken grouped by OwnCar')
plt.xlabel('Product Taken')
plt.ylabel('Age')
plt.show()
```

#### __Draw a heatmap among all numrical columns, the values should be displayed up to 3 decimal places__

```python
# Compute correlation matrix
corr_matrix = numerical_cols.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".3f",
    cmap='coolwarm',
    linewidths=0.5
)

plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```
