#### __Write code to remove M and k fomr the Size, convert the size to the Kilo bytes only__

```python
def removeMorK(x):
    if pd.isna(x):
        return np.nan

    x = str(x).strip().upper()

    if x.endswith('M'):
        return float(x[:-1]) * 1024  
    elif x.endswith('K'):
        return float(x[:-1])         
    else:
        return np.nan       

df['Size'] = df['Size'].apply(removeMorK)
```
